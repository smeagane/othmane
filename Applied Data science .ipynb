```python
stock_list=np.array([u'AAPL', u'MSFT', u'XOM', u'AMZN','BRK',u'JNJ', u'JPM','GOOG','GE','FB'])
```

```python
def get_ticker(tickers,startdate, enddate):
    array_tables=[pdr.get_data_yahoo(e, startdate, enddate) for e in tickers]
    return pd.concat(array_tables, keys=tickers, names=['Ticker', 'Date'])
```

```python
start=datetime.datetime(2014,1,1)
end=datetime.datetime(2015,1,1)
data=get_ticker(stock_list,startdate, enddate)
```
